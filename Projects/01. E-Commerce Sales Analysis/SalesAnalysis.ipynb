### 03. Clean the data

#### 03.01. Check how many NaNs there are

```python
nan_df = all_sales[all_sales.isna().any(axis=1)]
nan_df
```

#### 03.02. Drop All NaNs

```python
all_sales = all_sales.dropna(how='all')
all_sales.head(20)
```

#### 03.03. Find 'Or' and delete it

```python
##Checking the columns with 'Or' 
all_sales[all_sales.eq("Or").any(axis=1)]


## Removing 'Or'
all_sales = all_sales.drop(all_sales[all_sales.Month == 'Or'].index)
all_sales
```

```python
all_sales[all_sales.eq("Or").any(axis=1)]
```

### 04.Question-01: What was the best months for sales? How much was earned that month?

#### 04.01. Add month column

```python
all_sales['Month'] = 3
all_sales.head()
```

```python
all_sales['Month'] = all_sales['Order Date'].str[0:2].astype('int32')
all_sales.head()
```

#### 04.02. Convert str to INT in Month Column

```python
all_sales['Month'] = all_sales['Month'].astype('int32')
all_sales
```

#### 04.03. Convert other columns to correct type

```python
all_sales['Quantity Ordered'] = pd.to_numeric(all_sales['Quantity Ordered'])
all_sales['Price Each'] = pd.to_numeric(all_sales['Price Each']) 
```

#### 04.04. Add a sales column

```python
all_sales['Sales'] = all_sales['Quantity Ordered']*all_sales['Price Each']
all_sales
```

#### 04.05. Final Answer

```python
results = all_sales.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month no')
plt.show()
```
